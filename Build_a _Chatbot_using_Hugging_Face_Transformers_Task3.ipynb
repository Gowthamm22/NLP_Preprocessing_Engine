```python
#Step 2: Import Libraries

import torch                                                  #AutoTokenizer → converts text into tokens (numbers)
from transformers import AutoModelForCausalLM, AutoTokenizer  #AutoModelForCausalLM → generates responses using transformer model
```

```python
#Step 3: Load Pre-trained Model

model_name = "microsoft/DialoGPT-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)  #Loads DialoGPT, a conversational transformer model
```

```python
#Step 4: Start Chatbot

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")
```

```python
#Step 5: Initialize Chat Memory

chat_history_ids = None  #Stores previous conversation
```

```python
#Step 6: Start Continuous Conversation Loop \

while True:
    user_input = input("You: ").lower()
```

```python
#Step 7: Exit Condition

if user_input in ["exit", "quit"]:
    print("Chatbot: Chatbot ends the conversation")
    break
```

```python
for _ in range(10):   # allows 10 interactions
    user_input = input("You: ").lower()

    if user_input in ["exit", "quit"]:
        print("Chatbot: Chatbot ends the conversation")
        break

    elif "hello" in user_input:
        print("Chatbot: Hello! Nice to meet you. How can I assist you today?")

    elif "artificial intelligence" in user_input:
        print("Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving.")

    elif "python" in user_input:
        print("Chatbot: Python was created by Guido van Rossum and released in 1991.")

    elif "thank" in user_input:
        print("Chatbot: You're welcome! Feel free to ask more questions.")

    else:
        print("Chatbot: I'm still learning! Can you ask something else?")
```
